```python
import pandas as pd
weather = pd.read_csv('../content/Data/weatherAUS.csv')
```

```python
weather
```

```python
weather.columns
```

```python
weather.dropna(inplace=True)
```

```python
weather
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_features = ['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday']
encoded_features = OneHotEncoder(sparse_output=False)
encoded_features.set_output(transform='pandas')
encoded_features = encoded_features.fit_transform(weather[categorical_features])
```

```python
encoded_features
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_label = label_encoder.fit_transform(weather['RainTomorrow'])
```

```python
encoded_weather = encoded_features.assign(RainTomorrow=encoded_label)
```

```python
weather = weather.drop(columns=categorical_features+['RainTomorrow'])
```

```python
weather
```

```python
weather = weather.join(encoded_weather)
```

```python
weather
```

```python
from sklearn.model_selection import train_test_split
X = weather.drop(columns=['RainTomorrow'])
y = weather['RainTomorrow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import classification_report, f1_score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced', max_iter=2000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


print(classification_report(y_test, y_pred))
print('F1_score',f1_score(y_test, y_pred))
```
